```python
import seaborn as sns
import matplotlib as plt
```

```python
sns.set_theme(style="whitegrid")
```

```python
sender = df.loc[df['Message'] == 'Sender queried']
receiver = df.loc[df['Message'] == 'Receiver queried']
sender_update = df.loc[(df['Message'] == 'User updated') & (df['Name'] == 'Adam')]
receiver_update = df.loc[(df['Message'] == 'User updated') & (df['Name'] == 'Oliver')]
```

```python
def create_plot(array_of_df, pallette):
    for index, df  in enumerate(array_of_df):
        ax = sns.lineplot(data=df, x='Date', y='Message', marker='o', color=pallette[index])
        if index >= 2:
            for i, value in enumerate(df['Date']):
                ax.axvline(x=value, color=pallette[index], linestyle='--')
        for index, row in df.iterrows():
            ax.text(x=row['Date'], y=row['Message'], s=int(row['Amount']))
```

```python
sns.set(rc={'figure.figsize':(20,3)})
create_plot([sender, receiver, sender_update, receiver_update], ['Red', 'Green', 'Red', 'Green'])
for number in range(1, int(len(df)/4) + 1):
    sns.lineplot(data=df.loc[df['Number'] == number], x='Date', y='Message', marker='o', color='Grey')
```
